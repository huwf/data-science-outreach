```python
import pylab
import random

## First do a histogram of the true distribution, i.e.,
## uniform integers 1 to 6 inclusive
%pylab inline
uniformDist = [ 1, 2, 3, 4, 5, 6 ]

# print "Mean of 1--6 uniform is", pylab.mean(uniformDist)
# print "Variance =", pylab.var(uniformDist)
# print "SD =", pylab.std(uniformDist)

pylab.hist(uniformDist, bins= pylab.arange(0.5,6.6,1.0), normed=True)
pylab.axis([0.25,6.75,0,0.2])
pylab.xlabel('Throw of a single die',size="x-large")
pylab.ylabel('Probability',size="x-large")
pylab.title('Distribution of a Single Die')
pylab.savefig('oneDiePDF.png')
pylab.show()
```

```python
import pylab
import random
import math

fishData = [ 2.1, 2.4, 2.4, 2.4, 2.4, 2.6, 2.9, 3.2,
             3.2, 3.9, 4.5, 6.3, 8.2, 12.8, 23.5 ]


def geometricMean(listOfNumbers):
    """Calculates the geometric mean of a list of numbers."""
    product = 1
    for i in listOfNumbers:
        product *= i
    return product ** (1.0 / len(listOfNumbers))
```

```python

pylab.hist(fishData)
pylab.xlabel("Mass of the fish in kg",size='x-large')
pylab.ylabel("Count",size='x-large')
pylab.title("Histogram of Instances of Fish Mass")
pylab.savefig("fishHistogram1.png")
pylab.show()
```

```python



sampleSize = 100000

## Let's simulate the repeated throwing of a single six-sided die
singleDie = []
for i in range(sampleSize):
    newValue = random.randint(1,6)
    singleDie.append(newValue)

# print "Results for throwing a single die", sampleSize, "times."
# print "Mean of the sample =", pylab.mean(singleDie)
# print "Median of the sample =", pylab.median(singleDie)
# print "Standard deviation of the sample =", pylab.std(singleDie)
# print "Variance of the sample =", pylab.var(singleDie)
# print
# print

pylab.hist(singleDie, bins=[0.5,1.5,2.5,3.5,4.5,5.5,6.5] )
pylab.axis([0.1,6.9,0,18000])
pylab.title("Simulation of 100000 die throws")
pylab.xlabel('Number on die',size="x-large")
pylab.ylabel('Count',size="x-large")
pylab.savefig('singleDie.png')
pylab.show()
```

```python



## What about repeatedly throwing two dice and summing them?
twoDice = []
for i in range(sampleSize):
    newValue = random.randint(1,6) + random.randint(1,6)
    twoDice.append(newValue)

pylab.hist(twoDice, bins= pylab.arange(1.5,12.6,1.0))
pylab.axis([1.0,13.0,0,18000])
pylab.title("Simulation of the Scores of 100000 Throws of Two Dice")
pylab.xlabel('Total of two dice',size="x-large")
pylab.ylabel('Count',size="x-large")
pylab.savefig('twoDice.png')
pylab.show()
```
